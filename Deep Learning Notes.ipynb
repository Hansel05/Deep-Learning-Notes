## Deep Learning VS Machine Learning

**Deep Learning**
1. Performance of deep learning algorithm significantly increases as the quantity of data increases.
2. Appropriate hardware is required as deep learning algorithm undertakes complex matrix calculations.
3. Longer training time.
4. Faster prediction time.
5. Feature Selection is automatic.
6. The features extracted are not interpretable.

**Machine Learning**
1. Performance of machine learning algorithm stagnates after a certain mark as the quantity of data increases.
2. No hardware is required.
3. Shorter training time.
4. Prediction time depends.
5. Feature Selection is manual.
6. The features extracted are highly interpretable.

## Equation of line
$w_0 + w_1x_1 + w_2x_2 = 0$

Let $x_0 = 1$ such that,

$w_0x_0 + w_1x_1 + w_2x_2 = 0$

$\sum_{i=0}^{2} w_ix_i = 0$

$C + Ax + By = 0$
-
- Changes in the value of C => parallelly shifts the line.
- Changes in the value of A => rotates the line via y-axis.
- Changes in the value of B => rotates the line via x-axis.
___
$x_{new} = x_{old} - \eta Cooridnates$
___
Hence the formulated equation is:
- For correctly classified points => No changes to algorithm
- For incorrectly classified points => We need to make changes to algorithm

$w_{new} = w_{old} + \eta (y_i - \hat{y_i}) x_i$

```python
data = pd.DataFrame({
    'y': [1, 0, 1, 0],
    'y_hat': [1, 0, 0, 1],
    'y-y_hat': [0, 0, 1, -1]})
data
```

1. $w_{new} = w_{old} + \eta (0)$
2. $w_{new} = w_{old} + \eta (0)$
3. $w_{new} = w_{old} + \eta (1)$
4. $w_{new} = w_{old} + \eta (-1)$ = $w_{new} = w_{old} - \eta (1)$

```python
x, y = make_classification(n_features = 2, n_informative = 2, n_redundant = 0)
sns.scatterplot(x = x[:, 0], y = x[:, 1], c = y, cmap = 'rainbow')
plt.show()
```

```python
def perceptron(x, y, epoch):
    x = np.insert(x, 0, 1, axis = 1)
    weights = np.ones(x.shape[1])
    lr = 0.1
    
    for i in range(epoch):
        j = np.random.randint(1, 100)
        def step(z):
            return 1 if z > 0 else 0
        y_hat = step(np.dot(x[j], weights))
        weights = weights + lr * (y[j] - y_hat)*x[j]
        return weights[0], weights[1:]
```

```python
intercept, coef = perceptron(x, y, 500)
```

$Slope = \frac{-A}{B}$

$Intercept = \frac{-C}{B}$

```python
m = coef[0]/coef[1]
b = -intercept/coef[1]
print(f'm = {m}\nb = {b}')
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
sns.scatterplot(x = x[:, 0], y = x[:, 1], c = y, cmap = 'rainbow')
sns.lineplot(x = x_input, y = y_input, c = 'black')
plt.show()
```
